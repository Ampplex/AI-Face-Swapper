```python
import numpy as np 
import os 
import glob 
import cv2
import matplotlib. pyplot as plt
import numpy as np

import insightface
from insightface.app import FaceAnalysis
import cv2
```

```python
print('insightface:', insightface.__version__)  # Prints InsightFace version
print('numpy:', np.__version__)  # Prints NumPy version
```

Detecting Faces

```python
app = FaceAnalysis(name='buffalo_l')
app.prepare(ctx_id=0, det_size=(256, 256))
```

```python
swapper = insightface.model_zoo.get_model ('inswapper_128.onnx',
download=False)
```

```python
img1 = cv2.imread('elon_2.jpg')
img2 = cv2.imread('mark.webp')
```

```python
plt.imshow(img1[:,:,::-1])
plt.show()
```

```python
plt.imshow(img2[:,:,::-1])
plt.show()
```

# Image 1 to Image 2

```python
img1 = app.get(img1)
faces = app.get(img2)
img1 = img1[0]

res = img2.copy()

for face in faces:
    res = swapper.get(res, face, img1, paste_back=True)
```

```python
plt.imshow(res[:,:,::-1])
plt.show()
```

# Image 2 to Image 1

```python
img1 = cv2.imread('elon_2.jpg')
faces = app.get(img1)
img2 = app.get(img2)
img2 = img2[0]

res = img1.copy()

for face in faces:
    res = swapper.get(res, face, img2, paste_back=True)
```

```python
plt.imshow(res[:,:,::-1])
plt.show()
```
